```python
import numpy as np
import matplotlib.pyplot as plt
import v2_AGN_DataAndPlotting as adp
```

```python
def ToDict(zTable) : 
    # get redshifts 
    zTable_id, zTable_z = zTable.T # transpose to get columns 
    zTable_id = np.array(zTable_id, dtype=int) # get correct data type
    zTable_z  = np.array(zTable_z,  dtype=float)
    return {
        'ID' : zTable_id,
        'z'  : zTable_z
    }
```

```python
# open table with redshifts 
zTable_new = ToDict( np.array(adp.ReadFile(adp.path_cat+'\COSMOS_z_matches.csv')))
zTable_old = ToDict( np.array(adp.ReadFile(adp.path_cat+'\COSMOS_z_matches_old.csv')))
```

```python
plt.scatter(
    zTable_new['ID'],
    zTable_old['ID']
    )
plt.plot((-1,2E6),(-1,2E6), c='k')
plt.xlim((-1,1.2E6))
plt.ylim((-1,1.2E6))
plt.show()
```

```python
plt.scatter(
    zTable_new['z'],
    zTable_old['z'],
    )
plt.plot((-99,99),(-99,99), c='k')
plt.xlim((-1,11))
plt.ylim((-1,11))
plt.show()
```
