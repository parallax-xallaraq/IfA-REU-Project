```python
import scicomap as sc
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python
# custom colormaps 
red_cmap = mpl.colors.ListedColormap(['#6A040F','#9D0208','#D00000','#DC2F02','#E85D04','#F48C06','#FAA307','#FFBA08'])
blu_cmap = mpl.colors.ListedColormap(['#003E6E','#0461A7','#0699C6','#0FCBD1','#24E3C0','#47F08B','#94F76F','#BEF768'])
grn_cmap = mpl.colors.ListedColormap(['#054819','#075f07','#087708','#3da70c','#70d20f','#9ae227','#c5f240','#E5F250'])
```

```python

# returns the colormap corresponding to the key 
def GetCmap(key='turbo', n_ticks=9) :
    # red
    if( key=='red' or 
        key=='RED' or
        key=='r'
      ) : 
        return red_cmap
    # blue 
    if( key=='blu' or 
        key=='BLU' or
        key=='blue' or 
        key=='b'
      ) : 
        return blu_cmap
    # green
    if(
        key=='grn' or
        key=='GRN' or
        key=='green' or
        key=='g'
      ) :
        return grn_cmap
    # rainbow (default)
    return plt.cm.get_cmap('turbo', n_ticks-1)
```

# Assess Colormaps

```python
def ScatterUsingColors(cmapColors) :
    # example scatterplot
    plt.scatter(x=np.random.rand(100), y=np.random.rand(100), c=np.arange(100), cmap=mpl.colors.ListedColormap(cmapColors))
    plt.colorbar()
    plt.show()
```

```python
def ImproveCmap(cmap, lift=None, bitonic=True, diffuse=False) : 
    sc_map = sc.ScicoSequential(cmap=cmap)
    # assess current map 
    f=sc_map.assess_cmap(figsize=(22,10))
    f=sc_map.draw_example()
    # fixing the color map
    sc_map.unif_sym_cmap(lift=lift, bitonic=bitonic, diffuse=diffuse)
    # re-assess the color map after fixing it                     
    f=sc_map.assess_cmap(figsize=(22,10))
    f=sc_map.draw_example()
    # get colors
    correctedList = sc_map.get_mpl_color_map().colors
    print(correctedList)
    return(correctedList)   
```

## RED

```python
colors = ImproveCmap(GetCmap('r'),lift=None, bitonic=True, diffuse=False)
```

```python
rMap = GetCmap('b')
sc_map = sc.ScicoSequential(cmap=rMap)
f=sc_map.assess_cmap(figsize=(22,10))
f=sc_map.draw_example()
```
